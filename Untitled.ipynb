```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
link='https://raw.githubusercontent.com/AshishJangra27/Machine-Learning-with-Python-GFG/main/Feature%20Engineering/TSNE/bank_note_data.csv'
data=pd.read_csv(link)
```

```python
data.head()
```

```python
data['Class'].unique()
```

```python
data.info()
```

```python
features=list(data.columns)[:-1]
target=list(data.columns)[-1]
```

## Scaling Data

```python
from sklearn.preprocessing import StandardScaler
data_norm=data.copy()
sc=StandardScaler()
data_norm[features]=sc.fit_transform(data[features])
```

```python
data_norm
```

## TSNE

```python
from sklearn.manifold import TSNE
tsne=TSNE(learning_rate=500,n_components=2)
xtsne=tsne.fit_transform(data_norm[features])
```

```python
xtsne
```

```python
y=data[target]
y
```

 ## Visualisation

```python
plt.figure(figsize=(16,11))
plt.scatter(xtsne[np.where(y==0),0],xtsne[np.where(y==0),1],color='g',marker='^')
plt.scatter(xtsne[np.where(y==1),0],xtsne[np.where(y==1),1],color='r',marker='*')
```
